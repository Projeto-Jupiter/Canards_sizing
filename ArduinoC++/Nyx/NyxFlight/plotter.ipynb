```python
#Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib widget
```

```python
data = pd.read_csv('NyxData.csv')
data
```

```python
vector=[]
for i in data.index:
    vector.append(i*data["periodo(ms)"][i]*2/1000)
vector
```

```python
%matplotlib widget
plt.plot(vector, data['ax2_Raw']/2048) #Azul
plt.plot(vector, data['ay2_Raw']/2048) #Laranja
plt.plot(vector, data['az2_Raw']/2048) #Verde
plt.xlim(0,2)
```

```python
%matplotlib widget
plt.plot(vector, (np.pi/180)*data['gx2_Raw']/(16.4)) #Azul
plt.plot(vector, (np.pi/180)*data['gy2_Raw']/(16.4)) #Laranja
plt.plot(vector, (np.pi/180)*data['gz2_Raw']/(16.4)) #Verde
plt.xlim(0,20)
```

```python
%matplotlib widget
plt.plot(vector, data['lastCantAngle(°)'])
plt.xlim(2,5)
```

```python
%matplotlib widget
plt.plot(vector, (data['altitude(m)']/10)-620)
plt.xlim(0,12)
```
